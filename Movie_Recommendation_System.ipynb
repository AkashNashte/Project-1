```python
import ast

def convert(obj):
    # Parse the input string into a Python object
    parsed_obj = ast.literal_eval(obj)
    
    # Initialize an empty list to store names
    names = []
    
    # Iterate through each element in the parsed object
    for element in parsed_obj:
        # Extract the 'name' attribute and append it to the list
        names.append(element['name'])
    
    # Return the list of names
    return names
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
def convert3(obj):
    # Initialize an empty list to store the names of actors
    actor_names = []
    
    # Counter to keep track of the number of actors added
    counter = 0
    
    # Iterate over each actor in the movie object
    for actor in ast.literal_eval(obj):
        # Check if the counter is less than 3
        if counter < 3:
            # Add the actor's name to the list
            actor_names.append(actor['name'])
            # Increment the counter
            counter += 1
        else:
            # Break the loop if 3 actors are already added
            break
    
    # Return the list of top 3 actors
    return actor_names
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
movies.head()
```

```python
def fetch_director(obj):
    # Initialize an empty list to store the director's name
    director_name = []
    
    # Iterate over each crew member in the movie object
    for crew_member in ast.literal_eval(obj):
        # Check if the crew member's job is 'Director'
        if crew_member['job'] == 'Director':
            # Add the director's name to the list
            director_name.append(crew_member['name'])
            # Break the loop, as we found the director
            break
    
    # Return the list containing the director's name (or an empty list if director not found)
    return director_name
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
# Removing spaces between words in the 'genres' column
movies['genres'] = movies['genres'].apply(lambda x: [i.replace(" ", "") for i in x])

# Removing spaces between words in the 'keywords' column
movies['keywords'] = movies['keywords'].apply(lambda x: [i.replace(" ", "") for i in x])

# Removing spaces between words in the 'cast' column
movies['cast'] = movies['cast'].apply(lambda x: [i.replace(" ", "") for i in x])

# Removing spaces between words in the 'crew' column
movies['crew'] = movies['crew'].apply(lambda x: [i.replace(" ", "") for i in x])
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

### Text Vectorization

```python
# Importing CountVectorizer from scikit-learn
from sklearn.feature_extraction.text import CountVectorizer

# Initializing CountVectorizer with max_features=5000 and stop_words='english'
cv = CountVectorizer(max_features=5000, stop_words='english')

# Fitting CountVectorizer to the 'tags' column and transforming it into a numerical array
vectors = cv.fit_transform(new_df['tags']).toarray()

# Getting the feature names (most frequent 5000 words)
feature_names = cv.get_feature_names_out()
```

### Applying Stemming

```python
import nltk
```

```python
# Importing the PorterStemmer from the NLTK library
from nltk.stem import PorterStemmer
ps = PorterStemmer
```

```python
# Importing the PorterStemmer from the NLTK library
from nltk.stem import PorterStemmer

# Initializing the PorterStemmer
ps = PorterStemmer()

# Defining the stemming function
def stem(text):
    # Initialize an empty list to store the stemmed words
    stemmed_words = []

    # Iterate through each word in the text
    for word in text.split():
        # Apply stemming to the word and append the stemmed word to the list
        stemmed_word = ps.stem(word)
        stemmed_words.append(stemmed_word)

    # Join the stemmed words back into a single string and return it
    return " ".join(stemmed_words)

# Test the stem function with a sample text
stem_result = stem('In the 22nd century, a paraplegic Marine is dispatched to the moon Pandora on a unique mission, but becomes torn between following orders and protecting an alien civilization. Action Adventure Fantasy ScienceFiction cultureclash future spacewar spacecolony society spacetravel futuristic romance space alien tribe alienplanet cgi marine soldier battle loveaffair antiwar powerrelations mindandsoul 3d SamWorthington ZoeSaldana SigourneyWeaver JamesCameron')

print(stem_result)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

### Similarity Measures

```python
# Importing cosine_similarity from sklearn.metrics.pairwise
from sklearn.metrics.pairwise import cosine_similarity

# Calculating cosine similarity between vectors
similarity = cosine_similarity(vectors)
```

### Making the recommendation function

```python

def recommend(movie):

  movie_index = new_df[new_df['title'] == movie].index[0]

  distances = similarity[movie_index]

  movies_list = sorted(list(enumerate(distances)),reverse=True, key=lambda x:x[1])[1:6]


  for i in movies_list:

    print(new_df.iloc[i[0]].title)
```

### Recommended Movies
